CONTEO DE CADA PALABRA

```scala
def readFile(): Map[String, Int] ={
    val counter = scala.io.Source.fromFile("el_quijote.txt")
      .getLines
      .flatMap(_.split("\\W+"))
      .foldLeft(Map.empty[String, Int]){
        (count, word) => count + (word -> (count.getOrElse(word, 0) + 1))
      }
    
    return counter
  }

val counter = readFile
```

```scala
def printContent(counter : Map[String, Int]): Unit = {
    for ((k,v) <- counter) printf("%s : %d\n", k, v)
  }
```

```scala
printContent(counter)
```
