```python
import findspark
from pyspark.sql import SparkSession
```

```python
findspark.init()
spark= SparkSession.builder\
.master("local")\
.appName('El Quijote')\
.getOrCreate()
```

```python
libro = spark.read.text("el_quijote.txt")
```

CONTEO DE CADA PALABRA

```python
sc =  spark.sparkContext
in_text = sc.textFile('el_quijote.txt')

text_flat = in_text.flatMap(lambda x: x.split(' ')) \
.map(lambda x: (x,1)) \
.reduceByKey(lambda x,y : x+y)
```
